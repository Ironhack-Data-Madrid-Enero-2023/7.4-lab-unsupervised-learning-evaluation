```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore") 
```

```python
from sklearn import datasets
#from sklearn.preprocessing import StandardScaler

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
#X = StandardScaler().fit_transform(X)
y = pd.Series(data["target"])

X.head()
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#Preprocess the Data Set
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# Train the KMeans Clustering Model
kmeans = KMeans(n_clusters=8, random_state=42)
kmeans.fit(scaled_data)

#Compute the Silhouette Score
score = silhouette_score(scaled_data, kmeans.labels_)
print("Silhouette Score:", score)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
#Train the KMeans Clustering Model

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(scaled_data)

#Compute the Silhouette Score

score = silhouette_score(scaled_data, kmeans.labels_)
print("Silhouette Score:", score)
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

model = KMeans(random_state=42)
visualizer = KElbowVisualizer(model, k=(1, 11), metric='distortion')
visualizer.fit(scaled_data)
visualizer.show()
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
# create lists to store distortion and silhouette scores for different k values
distortions = []
silhouette_scores = []

# test different k values for KMeans
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    distortions.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# plot the elbow curve for KMeans
plt.plot(range(2, 10), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('KMeans elbow curve')
plt.show()

# plot the silhouette scores for KMeans
plt.plot(range(2, 10), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('KMeans silhouette scores')
plt.show()

# create lists to store silhouette scores for different k values for AgglomerativeClustering
agg_silhouette_scores = []

# test different k values for AgglomerativeClustering
for k in range(2, 10):
    agg = AgglomerativeClustering(n_clusters=k)
    agg.fit(X)
    agg_silhouette_scores.append(silhouette_score(X, agg.labels_))

# plot the silhouette scores for AgglomerativeClustering
plt.plot(range(2, 10), agg_silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.title('AgglomerativeClustering silhouette scores')
plt.show()
```

Use the model and number of clusters that returned the best result above.

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# plot the first two principal components with point colors determined by cluster
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=agg.labels_)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA plot with colored clusters')
plt.show()
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# create a list of perplexities to try
perplexities = [5, 10, 20, 30, 50, 100]

# create subplots for each perplexity
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
axs = axs.flatten()

# loop through the perplexities and plot the t-SNE embeddings with colored clusters
for i, perp in enumerate(perplexities):
    tsne = TSNE(n_components=2, perplexity=perp)
    X_tsne = tsne.fit_transform(X)
    axs[i].scatter(X_tsne[:, 0], X_tsne[:, 1], c=agg.labels_)
    axs[i].set_title(f"t-SNE plot with perplexity={perp}")

plt.tight_layout()
plt.show()
```
